```python
# calculate % of churn within each SeniorCitizen group
total_counts = df.groupby("SeniorCitizen")['Churn'].value_counts(normalize=True).unstack() * 100

# plot
fig, ax = plt.subplots(figsize=(6,6))

# stacked bar plot
total_counts.plot(kind="bar", stacked=True, ax=ax, color=["#1f77b4", "#ff7f0e"])

# add % labels
for p in ax.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy()
    if height > 0:
        ax.text(x + width / 2, y + height / 2, f'{height:.1f}%', ha='center', va='center', fontsize=10)

# labels and title
plt.title("Churn by SeniorCitizen (Stacked % Bar Chart)")
plt.xlabel("SeniorCitizen")
plt.ylabel("Percentage (%)")
plt.xticks(rotation=0)
plt.legend(title='Churn', loc='upper right') #,bbox_to_anchor=(0.9,0.9) # customize legend location
plt.tight_layout()
plt.show()
```

```python

# Example: assuming your data is in a DataFrame called df
# df = pd.read_csv('your_file.csv')

columns = [
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]

# Set up the plot grid
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(18,20))  # 3x3 grid for 9 plots
axes = axes.flatten()  # flatten to easily iterate

# Loop through each column and plot
for i, col in enumerate(columns):
    sns.countplot(data=df, x=col, ax=axes[i])
    axes[i].set_title(f'Count Plot of {col}')
    axes[i].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

```python
# Example: assuming your data is in a DataFrame called df
# df = pd.read_csv('your_file.csv')

columns = [
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]

# Set up the plot grid
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(18,20))  # 3x3 grid for 9 plots
axes = axes.flatten()  # flatten to easily iterate

# Loop through each column and plot
for i, col in enumerate(columns):
    sns.countplot(data=df, x=col, ax=axes[i],hue=df["Churn"])
    axes[i].set_title(f'Count Plot of {col}')
    axes[i].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

The visualizations present count plots showing customer churn (yes or no) across various service features such as phoneservice,Multiplelines, and onlinesecurity. The majority of customers who do not churn tend to have services like onlinebackup. techsupport, and streamingTV, churn rates are noticeably higher when these services are not used or are unavailable. The charts emphasize the relationship between service usage and customer churn.
